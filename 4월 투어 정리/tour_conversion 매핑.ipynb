```python
import pandas as pd
import json
import os

# Excel 파일 경로
excel_file = r'C:\Users\ch.kang\Desktop\4월투어\ga_20240419.xlsx'

# 만든 고유 아이디 매핑 파일 경로
mapping_file = r'C:\Users\ch.kang\OneDrive - 클리 주식회사\문서 - 클리주식회사\000_Routine_마이세컨플레이스\02. IT Platform&Data\01_콘텐츠성과측정\02_ConvertedData\id_mapping.json'

# 데이터프레임 로드
df = pd.read_excel(excel_file, dtype=str)

# 고유 아이디 매핑 파일 로드
with open(mapping_file, 'r') as f:
    id_mapping = json.load(f)

# UniqueIDGenerator 클래스
class UniqueIDGenerator:
    def __init__(self, id_mapping):
        self.id_mapping = id_mapping

    def get_unique_id(self, username):
        return self.id_mapping.get(username, None)

# UniqueIDGenerator 인스턴스 생성
generator = UniqueIDGenerator(id_mapping)

# 고유 아이디 생성 및 데이터프레임에 추가
df['고유아이디'] = df['아이디'].apply(generator.get_unique_id)

# 결과 확인
print(df)
```

```python
df_pageview = df[df['이벤트명'] == 'page_view']
df_pageview.head(5)
```

```python
# conversion에 따라 고유 아이디를 표로 정리합니다.
conversion_ids = df_pageview.groupby('conversion')['고유아이디'].apply(list).reset_index()
only_conversion_ids = conversion_ids.drop(0)
expanded_ids = only_conversion_ids.explode('고유아이디')
```

```python
expanded_ids
# conversion을 기준으로 그룹화하여 '고유아이디'를 리스트로 묶기
grouped_ids = expanded_ids.groupby('conversion')['고유아이디'].apply(list).reset_index()
grouped_ids['고유아이디'] = grouped_ids['고유아이디'].astype(str)
grouped_ids['고유아이디'] = grouped_ids['고유아이디'].str.replace('[', '').str.replace(']', '')

```

```python
df_expanded = grouped_ids['고유아이디'].str.split(',', expand=True)
df_result = pd.concat([grouped_ids, df_expanded], axis=1)
df_result = df_result.drop(columns=['고유아이디'])

```

```python
conversion_df=df_result.T
# 1번째 행을 새로운 열 헤더로 설정
new_header = conversion_df.iloc[0]
conversion_df = conversion_df[1:]
conversion_df = conversion_df.set_axis(new_header, axis=1)
conversion_df.loc['Count'] = conversion_df.count()
conversion_df = conversion_df.reindex(index=['Count'] + conversion_df.index[:-1].tolist())
```

```python
output_path = os.path.join(os.path.dirname(excel_file), 'conversion_' + os.path.basename(excel_file))

with pd.ExcelWriter(output_path) as writer:
    # df를 첫 번째 시트로 저장
    df.to_excel(writer, sheet_name='rawdata', index=False)
    
    # conversion_ids를 두 번째 시트로 저장
    conversion_df.to_excel(writer, sheet_name='conversion_ids', index=False)
```
