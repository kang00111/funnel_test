```python
# 폴더 내의 파일 목록 가져오기
file_list = os.listdir(convert_path)
ga_files = [file for file in file_list if file.startswith('ga') and file.endswith('newID.xlsx')]

ga_files
```

```python
# 'ga'로 시작하는 전체 파일 필터링
combined_data = pd.DataFrame()
for file in ga_files:
    file_path = os.path.join(convert_path, file)
    data = pd.read_excel(file_path, dtype = str)  # 엑셀 파일 읽기
    combined_data = pd.concat([combined_data, data], ignore_index=True)  # 데이터 합치기

# 여기서 combined_data를 사용하여 필요한 작업을 수행합니다.
```

```python
id_name_mapping = {
    'g-030446': '안성우',
    'g-030397': '원일한',
    'g-030172': '이호',
    'g-030164': '민병제',
    'g-030093': '이해주',
    'g-019403': '이재용',
    'g-022098': '성선영'
}

target_ids = ['g-030446', 'g-030397', 'g-030172', 'g-030164', 'g-030093', 'g-019403', 'g-022098']
filtered_data = combined_data[combined_data['고유아이디'].isin(target_ids)]
filtered_data['이름'] = filtered_data['고유아이디'].map(id_name_mapping)
filtered_data.head(5)
```

```python
output_file_path = os.path.join(request_path, "new_id별 행동_미등기양성화구매자.xlsx")
filtered_data.to_excel(output_file_path, index=False)
```
